**Exploratory Data Analysis (Graphs)**

1. Distribution of House Prices

```python
plt.figure(figsize=(10,5))
plt.hist(df['price'], bins=50)
plt.title("Distribution of House Prices in Mumbai")
plt.xlabel("Price (in Rupees)")
plt.ylabel("Number of Properties")
plt.show()
```

2 Distribution of Property Area

```python
plt.figure(figsize=(10,5))
plt.hist(df['area'], bins=50)
plt.title("Distribution of Property Area")
plt.xlabel("Area (sq ft)")
plt.ylabel("Number of Properties")
plt.show()
```

3 Price vs Area (Scatter Plot)

```python
plt.figure(figsize=(10,6))
plt.scatter(df['area'], df['price'], alpha=0.3)
plt.title("Price vs Area")
plt.xlabel("Area (sq ft)")
plt.ylabel("Price")
plt.show()
```

4 Correlation Heatmap

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])

plt.figure(figsize=(10,8))
sns.heatmap(numeric_df.corr(), annot=True, fmt=".2f")
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```

**Top 10 Most Expensive Localities in Mumbai**

* This will show which areas have the highest average property price.

```python
top_localities = df.groupby("locality")["price"].mean().sort_values(ascending=False).head(10)
top_localities
```

```python
plt.figure(figsize=(12,6))
top_localities.plot(kind='bar')
plt.title("Top 10 Most Expensive Localities in Mumbai")
plt.xlabel("Locality")
plt.ylabel("Average Price (in Rupees)")
plt.xticks(rotation=45)
plt.show()
```

**Price vs Number of Bedrooms (BHK)**

* This shows how the average price changes as the number of bedrooms increases.

```python
bhk_price = df.groupby("bedroom_num")["price"].mean()

plt.figure(figsize=(10,5))
bhk_price.plot(kind='bar')
plt.title("Average Price by Number of Bedrooms")
plt.xlabel("Number of Bedrooms")
plt.ylabel("Average Price (in Rupees)")
plt.xticks(rotation=0)
plt.show()
```

**Price per Sqft â€” Top Expensive Localities**

```python
top_pps = df.groupby("locality")["price_per_sqft"].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
top_pps.plot(kind='bar')
plt.title("Top 10 Localities by Price per Sq Ft")
plt.xlabel("Locality")
plt.ylabel("Average Price per Sq Ft (INR)")
plt.xticks(rotation=45)
plt.show()
```
